##  Objective:
To preprocess a raw dataset by handling missing values, encoding categorical variables, scaling features, and splitting data.

##  Importing Libraries

```python
import pandas as pd
import numpy as np
```

##  Loading Dataset

```python
df = pd.read_csv('/content/1) iris.csv')
df.head()
```

##  Data Understanding

```python
# Basic information
df.info()

# Statistical summary
df.describe()

# Check missing values
df.isnull().sum()
```

##  Handling Missing Values

```python
# Fill missing numerical values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

##  Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical column into numeric
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])

df.head()
```

##  Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('species', axis=1)
y = df['species']

# Apply scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Convert scaled data into DataFrame (for better display)
scaled_df = pd.DataFrame(
    X,
    columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
)

# Display first 5 rows of scaled data
scaled_df.head()
```

##  Splitting Data

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Check shape of split data
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```
